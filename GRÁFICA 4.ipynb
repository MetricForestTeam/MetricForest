```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')


# Porcentaje de árboles con caracteristicas deseables y no deseables
x = [481,41]
colors = plt.get_cmap('GnBu')(np.linspace(0.2, 0.7, len(x)))

nombres = ["No deseable","Deseable"]
plt.pie(x, labels=nombres)
plt.axis("equal")
plt.title('Árboles con caracteristicas deseables')
plt.show()
```
