```python
#Especies más abundantes
##Encino(228), Roble(179)

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
x = np.linspace(-179, 228, 100)
y = np.linspace(-179, 228, 100)
x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + 2*y**2)
ax.contourf(x, y, z, cmap='Greens')

plt.title('Especies más abundantes')


plt.contourf(x, y, z, cmap='Greens')
plt.colorbar();

plt.rcParams['figure.figsize'] = [4,3]
plt.show()
```
