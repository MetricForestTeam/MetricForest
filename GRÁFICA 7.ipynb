```python
import pandas as pd
from matplotlib import pyplot as plt
prueba=pd.read_csv("prueba.csv")
prueba
Encino=prueba[prueba.Especie=="Encino"]
Encino
Roble=prueba[prueba.Especie=="Roble"]
Roble
#Caracteristicas deseables según especie
plt.plot(Encino.Sector, Encino.LD)
plt.plot(Roble.Sector, Roble.LD)
plt.title('Caracteristicas deseables según especie')
plt.legend(["Encino", "Roble"])
plt.xlabel("Sector")
plt.ylabel("LD")
plt.rcParams['figure.figsize'] = [3,3]

plt.show()
```
