```python
#Comparación de diametro con altura según parcelas
df = pd.DataFrame({'Parcela':['1', '2', '3', '4', '5'], 
                   'DAP':[16.08, 15.76, 11.49, 16.39, 18.16], 
                   'HT':[5.39, 5.44, 7.79, 7.61, 8.422]})

df = df.set_index('Parcela')
fig, ax = plt.subplots()
df.plot(ax = ax) 
plt.show()

plt.rcParams['figure.figsize'] = [3,3]
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-pastel')

#Comparación de diametro con altura según parcelas
fig, ax = plt.subplots(2, 2, sharey = True)
parcela = ['1', '2', '3', '4', '5']
parámetro = {'DAP':[16.08, 15.76, 11.49, 16.39, 18.16], 'HT':[5.39, 5.44, 7.79, 7.61, 8.422]}
ax[0, 0].plot(parcela, parámetro['DAP'], color = 'tab:green')
ax[0, 1].plot(parcela, parámetro['HT'], color = 'tab:olive')
ax[1, 0].bar(parcela, parámetro['DAP'], color = 'tab:green')
ax[1, 1].bar(parcela, parámetro['HT'], color = 'tab:olive')

plt.show()

plt.rcParams['figure.figsize'] = [4,3]
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')


# Porcentaje de árboles con caracteristicas deseables y no deseables
x = [481,41]
colors = plt.get_cmap('GnBu')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=2, center=(2,2),
       wedgeprops={"linewidth": 3, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 4), xticks=np.arange(1, 4),
       ylim=(0, 4), yticks=np.arange(1, 4))

plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make data
X, Y = np.meshgrid(np.linspace(-179, 228, 526), np.linspace(-179, 228, 526))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(Z.min(), Z.max(), 7)

# plot
fig, ax = plt.subplots()

ax.contourf(X, Y, Z, levels=levels)

plt.show()
```
