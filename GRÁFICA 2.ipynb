```python
import pandas as pd 
import matplotlib.pyplot as plt

#Comparación de diametro con altura según parcelas
df = pd.DataFrame({'Parcela':['1', '2', '3', '4', '5'], 
                   'DAP':[16.08, 15.76, 11.49, 16.39, 18.16], 
                   'HT':[5.39, 5.44, 7.79, 7.61, 8.422]})

df = df.set_index('Parcela')
fig, ax = plt.subplots()
df.plot(ax = ax) 
plt.show()

plt.rcParams['figure.figsize'] = [3,3]
```
