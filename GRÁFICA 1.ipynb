```python
#Grfica de comparacin de la abundancia de especie por sector.

import pandas as pd
from matplotlib import pyplot as plt


prueba=pd.read_csv("prueba.csv")
prueba

grafica = prueba.groupby(["Especie", "Sector"], sort=False,)["DAP"].max()
grafica.unstack().T.plot(kind="bar", subplots=False, layout=(2,2), figsize=(20,10), fontsize=(25))

plt.title("Especie vs Sector") 
```
