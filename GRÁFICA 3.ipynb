```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-pastel')

#Comparación de diametro con altura según parcelas
fig, ax = plt.subplots(2, 2, sharey = True)
parcela = ['1', '2', '3', '4', '5']
parámetro = {'DAP':[16.08, 15.76, 11.49, 16.39, 18.16], 'HT':[5.39, 5.44, 7.79, 7.61, 8.422]}
ax[0, 0].plot(parcela, parámetro['DAP'], color = 'tab:green')
ax[0, 1].plot(parcela, parámetro['HT'], color = 'tab:olive')
ax[1, 0].bar(parcela, parámetro['DAP'], color = 'tab:green')
ax[1, 1].bar(parcela, parámetro['HT'], color = 'tab:olive')

plt.show()

plt.rcParams['figure.figsize'] = [4,3]
```
